```python
import numpy as np
import matplotlib.pyplot as plt
```

Let's create some 2D array with random numbers using the numpy `random` module. We can first check how it works:

```python
np.random?
```

I would like normaly distributed value, so let's check random.randn

```python
np.random.randn?
```

Let's create an array of normaly distributed random numbers that have 20 rows and 30 columns

```python
test_array = np.random.randn(20, 30)
```

To make sure we did this right, let's check the shape of the array

```python
test_array.shape
```

### Visualizing 2D data 
Now, let's visualize this data using the pyplot function `imshow`

```python
image = plt.imshow(test_array)
plt.show()
```

The value on each row and column is displayed as a heatmap. Note that the graph has 20 values on the y-axis (rows) and 30 values on x-axis (columns). Blue pixels in this heat map represent low values, while yellow pixels represent high values. 


We can also add a colorbar to have a better idea of the values.

```python
image = plt.imshow(test_array)
plt.colorbar()
plt.show()
```

### Looking at distributions
We have created random data that is supposed to be normaly distributed. We can check the distribution of a certain variable by looking at a histogram.

```python
histogram = plt.hist(test_array.flatten())
```

### Visualizing 1D data

Now, let's assume this random data that we created represents the anomaly in the mean price of 20 (rows) different cryptocurrencies (Dogecoin included!) over the last 30 days (columns). If we want to look at the average variation in all cryptocurrencies during the last 30 days, we can plot:

```python
ave_price = np.mean(test_array, axis=0)
figure = plt.plot(ave_price)
plt.show()
```

We can make this a bit better by adding information to our plot

```python
ave_price = np.mean(test_array, axis=0)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111)
ax.plot(ave_price, color='red')
ax.set_ylabel('Price Anomaly', fontsize=14)
ax.set_xlabel('Days', fontsize=14)
plt.show()
```
